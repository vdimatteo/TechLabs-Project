```python

import pandas as pd

df = pd.read_csv("confirmed.csv")
```

```python
df.rename(columns={'Country/Region':'Datetime'}, inplace=True)

df.drop(['Lat', 'Long'],axis=1, inplace=True)



df.head()
```

```python
Countries = df.groupby("Datetime").sum()
Countries.head()

```

```python
transform = Countries.T

transform.head()
```
