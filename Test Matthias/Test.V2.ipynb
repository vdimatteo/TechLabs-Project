```python

import pandas as pd

df = pd.read_csv("confirmed.csv")
```

```python

df.drop(['Lat', 'Long'],axis=1, inplace=True)



df.head()
```

```python
Countries = df.groupby("Country/Region").sum()
Countries.head()

```

```python
transform = Countries.T

transform.head()
```

```python
transform.index = pd.PeriodIndex(transform.index, freq='D')
transform.index 
```

```python

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
country = input("Please enter country : ")


transform[country].plot()
```
